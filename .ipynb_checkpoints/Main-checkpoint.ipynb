**Label Encoding**

```python
#creating a copy of original dataset 
new_df=df.copy()
new_df.head()
```

```python
# Checking how many attributes are of type object

objList = new_df.select_dtypes(include = "object").columns
print (objList)
```

```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in objList:
    new_df[feat] = le.fit_transform(new_df[feat].astype(str))

print (new_df.info())
```

**Multicollinearity**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(df):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = df.columns
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

    return(vif)

calc_vif(new_df)
```

```python
# Handling multicollinearity
new_df['origin_txn_amount'] = new_df.apply(lambda x: x['oldbalanceOrg'] - x['newbalanceOrig'],axis=1)
new_df['dest_txn_amount'] = new_df.apply(lambda x: x['oldbalanceDest'] - x['newbalanceDest'],axis=1)
new_df['txn_path'] = new_df.apply(lambda x: x['nameOrig'] + x['nameDest'],axis=1)

#Dropping columns
new_df = new_df.drop(['oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest','step','nameOrig','nameDest'],axis=1)

calc_vif(new_df)
```

```python
# analyze the heatmap now
corr=new_df.corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True,cmap="Blues")
```

# Model Building using RandomForest Classifier

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import itertools
from collections import Counter
import sklearn.metrics as metrics
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

Normalization

```python
scaler = StandardScaler()
new_df["NormalizedAmount"] = scaler.fit_transform(new_df["amount"].values.reshape(-1, 1))
new_df.drop(["amount"], inplace= True, axis= 1)

Y = new_df["isFraud"]
X = new_df.drop(["isFraud"], axis= 1)
```

```python
# Split the data
(X_train, X_test, Y_train, Y_test) = train_test_split(X, Y, test_size= 0.3, random_state= 42)

print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

# Model Training

```python
# RANDOM FOREST

random_forest = RandomForestClassifier(n_estimators= 100)
random_forest.fit(X_train, Y_train)

Y_pred_rf = random_forest.predict(X_test)
random_forest_score = random_forest.score(X_test, Y_test) * 100
```

# Evaluation

```python
print("TP,FP,TN,FN - Random Forest")
tn, fp, fn, tp = confusion_matrix(Y_test, Y_pred_rf).ravel()
print(f'True Positives: {tp}')
print(f'False Positives: {fp}')
print(f'True Negatives: {tn}')
print(f'False Negatives: {fn}')


print("Random Forest Score: ", random_forest_score)
```

```python
confusion_matrix_rf = confusion_matrix(Y_test, Y_pred_rf.round())
print("Confusion Matrix - Random Forest")
print(confusion_matrix_rf)
```

```python
classification_report_rf = classification_report(Y_test, Y_pred_rf)
print("Classification Report - Random Forest")
print(classification_report_rf)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_rf)
disp.plot()
plt.title('Confusion Matrix - RF')
plt.show()
```

```python
#plotting the ROC curve
fpr, tpr, threshold = metrics.roc_curve(Y_test, Y_pred_rf)
roc_auc = metrics.auc(fpr, tpr)

plt.title('ROC - RF')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

**1)Data cleaning including missing values, outliers and multi-collinearity.**

Ans. In my project, I ensured clean data before modeling. First, I checked for missing values but found none, so no imputation was needed. For multicollinearity, I used VIF to identify highly correlated balance features. Instead of dropping them blindly, I engineered new features like transaction amount differences and removed redundant variables. This gave me a cleaner, more robust dataset for fraud detection.


**2)Describe your fraud detection model in elaboration.**

Ans. I built a fraud detection model using Random Forest on transactional data. After performing EDA, feature engineering, label encoding, and normalization, I trained the model and evaluated it using confusion matrix, classification report, and ROC-AUC. The model achieved high accuracy and good fraud detection performance, especially in identifying fraudulent transfers and cash-outs.


**3)How did you select variables to be included in the model?**

Ans. I selected variables based on EDA insights, correlation analysis, and multicollinearity checks. Categorical variables like type were label-encoded, and only meaningful, less-collinear variables were kept for the model.
